```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import  seaborn as sns
import scipy.stats as stats

pd.pandas.set_option('display.max_columns',None)
```

```python
data = pd.read_csv('train.csv')
```

```python
data.drop('Id',axis=1,inplace=True)
```

```python
target = 'SalePrice'

data[target].hist(bins=50,density=True)
plt.xlabel('Sale Price')
plt.ylabel('Number of Houses')
plt.show()
```

```python
np.log(data[target]).hist(bins=50,density=True)
plt.xlabel('Log of Sale Price')
plt.ylabel('Number of Houses')
plt.show()
```

```python
cat_vars = list(data.select_dtypes(include='object').columns) + ['MSSubClass']
len(cat_vars)
```

```python
data[cat_vars] = data[cat_vars].astype('object')
```

```python
num_vars = data.iloc[:,:-1].select_dtypes(include=['int','float']).columns.to_list()
```

```python
num_vars
```

```python
var_with_nan = [var for var in data.columns if data[var].isna().sum() > 0]

data[var_with_nan].isna().mean().sort_values(ascending=False)
```

```python
data[var_with_nan].isna().mean().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Data columns with missing values')
plt.ylabel('Percentage of missing values')
plt.show()
```
